```r
#Import the experimental dataset (the file contains all experiments performed up to 4GPa)
exp_cpx<- read.delim("CPX_experiments.txt")
```

```r
#Al2O3 in cpx seems to show a good relationships with pressure
plot(exp_cpx$Al2O3.cpx, exp_cpx$P_GPa, pch=19, col="red", ylab="Experimental Pressure [GPa]", xlab="Al2O3 in cpx")
```

Let's try to fit a linear regression through the data using only Al2O3 as independent variable

```r
Y <- exp_cpx$P_GPa
X <- exp_cpx$Al2O3.cpx
xlab<- "Al2O3 [wt.%]"
ylab<- "Experimental Pressure [GPa]"

model.data <- data.frame(Y,X)
linearMod <- lm(Y ~ X, data=model.data)
summary(linearMod)


plot(X,Y,pch=21,col="grey",bg="black",xlab=xlab,ylab=ylab)
lines(sort(X), fitted(linearMod)[order(X)], col='red', type='l',lwd=2)
```

What about a non linear model?

```r
nonlinearMod <- lm(Y ~ I(X^0.5), data=model.data)
print(nonlinearMod)
summary(nonlinearMod)

plot(X,Y,pch=21,col="grey",bg="black",xlab=xlab,ylab=ylab)
lines(sort(X), fitted(nonlinearMod)[order(X)], col='red', type='l',lwd=2)
```
